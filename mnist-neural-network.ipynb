```python
import numpy as np 
import pandas as pd 

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical, plot_model
```

```python
dataset = pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
X = dataset.iloc[:,1:]
y = to_categorical(dataset.iloc[:,0])
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.1)
```

```python
model = Sequential()
model.add(Dense(12, input_shape=(784,), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(10, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=5, batch_size=10)
```

```python
y_pred = model.predict(X_test)
pred = list()
for i in range(len(y_pred)):
    pred.append(np.argmax(y_pred[i]))
test = list()
for i in range(len(y_test)):
    test.append(np.argmax(y_test[i]))
```

```python
dt = pd.DataFrame(columns=['ImageId','Label'])
dt['Label'] = np.array(pred).tolist()
```

```python
dt.index = pd.RangeIndex(start=0, stop=0+len(dt), step=1)
dt ['ImageId']=dt.index
dt.to_csv('test.csv')
dt
```

```python
from sklearn.metrics import accuracy_score
a = accuracy_score(pred, test)
print('Accuracy is:', a*100, '%')
```
